```python
import torch
from torch import nn
from torchvision import datasets
from torch.utils.data import DataLoader
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
```

```python
train_data = datasets.CIFAR10(
    root = "data",
    train = True,
    download = True,
    transform = ToTensor()
)
test_data = datasets.CIFAR10(
    root = "data",
    train = False,
    download = True,
    transform = ToTensor()
)
```

```python
train_dataloader = DataLoader(train_data, batch_size = 123, shuffle = True)
test_dataloader = DataLoader(test_data, batch_size = 123, shuffle = True)
```

```python
class linear_classifier:
  def __init__(self, in_feature , out_feature):
    self.weight = torch.rand((in_feature, out_feature)).requires_grad_()
    self.bias = torch.zeros(out_feature).requires_grad_()
  def forward(self, X):
    self.output = torch.matmul(X, self.weight) + self.bias
```

```python
class support_vector_mashine:
  def calculate(self,layer_output, label , margin):
    self.elements = layer_output.shape[0]
    self.loss = torch.sum(torch.max(layer_output - layer_output[label] + margin, torch.tensor(0)))/self.elements
```

```python
class optimizer:
  def __init__(self, lr = 1., decay = 1e-3):
    self.learning_rate = lr
    self.current_learning_rate = lr
    self.decay = decay
    self.step = 0
  def before_update_params(self):
    self.current_learning_rate = self.learning_rate * (1./1+self.decay * self.step)
  def update_params(self, weight, bias):
    weight = -weight + weight * self.current_learning_rate
    bias = -bias + bias * self.current_learning_rate
  def after_update_params(self):
    self.step += 1
```

```python
class RELU:
  def forward(self,layer_output):
    layer_output[layer_output < 0] = 0
    self.output = layer_output
```

```python
l1 = linear_classifier(3*32*32, 100)
a1 = RELU()
l2 = linear_classifier(100, 100)
a2 = RELU()
l3 = linear_classifier(100,10)
loss = support_vector_mashine()
optm = optimizer(lr = 0.5)
for i in range(10):
  for img, lbl in train_dataloader:
    img = img.reshape(-1,3*32*32)/255
    l1.forward(img)
    a1.forward(l1.output)
    l2.forward(a1.output)
    a2.forward(l2.output)
    l3.forward(l2.output)
    loss.calculate(l3.output,lbl, 10)
    print(f'Loss = {loss.loss:.4f}')
    loss.loss.backward()
    optm.before_update_params()
    optm.update_params(l1.weight, l1.bias)
    optm.update_params(l2.weight, l2.bias)
    optm.update_params(l3.weight, l3.bias)
    optm.after_update_params()
```
